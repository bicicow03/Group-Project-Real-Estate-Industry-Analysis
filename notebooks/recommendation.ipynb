```python
import pandas as pd 

houses = pd.read_csv("../data/outputs/houses_growth.csv")

apartments = pd.read_csv("../data/outputs/apartments_growth.csv")

overall = pd.read_csv("../data/outputs/overall_growth.csv")

import pandas as pd 

affordable = pd.read_csv("../data/outputs/affordable.csv")
livable = pd.read_csv("../data/outputs/livable.csv")

rentals = pd.read_csv("../data/processed/real_estate/vic_rentals_all_enriched.csv")

drop = [column for column in rentals.columns if column not in ['suburb', 'postcode']]

sub = rentals.drop(columns=drop)

suburb = sub.drop_duplicates()

most_affordable = affordable.merge(suburb, on='postcode', how='left')

most_livable = livable.merge(suburb, on='postcode', how='left')



```

```python


apartments['unit_avg_price_today'] = (apartments['unit_1bed_Rent_today'] + apartments['unit_2bed_Rent_today'] + apartments['unit_3bed_Rent_today']) /3

apartments['unit_avg_price_5yr'] = (apartments['unit_3bed_2030_avg'] + apartments['unit_2bed_2030_avg'] + apartments['unit_1bed_2030_avg']) /3

apartments['unit_5yr_change(%)'] = (apartments['unit_avg_price_5yr'] - apartments['unit_avg_price_today']) / apartments['unit_avg_price_today'] *100

apartments.sort_values(by='unit_5yr_change(%)', ascending=False)

keep = ['Suburb', 'apartment_price', 'unit_avg_price_today', 'unit_avg_price_5yr', 'unit_5yr_change(%)', 'apartment_ALL_avg_price_change(%)']

drop = [col for col in apartments.columns if col not in keep]

apt = apartments.drop(columns=drop)

apt.sort_values(by='unit_5yr_change(%)', ascending=False).head()
```

```python
houses['house_avg_price_today'] = (houses['house_2bed_Rent_today'] + houses['house_3bed_Rent_today'] + houses['house_4bed_Rent_today']) /3

houses['house_avg_price_5yr'] = (houses['house_4bed_2030_avg'] + houses['house_3bed_2030_avg'] + houses['house_2bed_2030_avg']) /3

houses['house_5yr_change(%)'] = (houses['house_avg_price_5yr'] - houses['house_avg_price_today'])/houses['house_avg_price_today'] *100

keep_h = ['Suburb', 'House_price', 'house_avg_price_today', 'house_avg_price_5yr', 'house_5yr_change(%)', 'House_ALL_avg_price_change(%)']

drop_h = [col for col in houses.columns if col not in keep_h]

hou = houses.drop(columns=drop_h)

hou.sort_values(by='house_5yr_change(%)', ascending=False).head()
```

```python
all = hou.merge(apt, on='Suburb', how='left')

all.head()
```

```python

apartments['5yr_income_1bed'] = (apartments['unit_1bed_2026_avg'] + apartments['unit_1bed_2027_avg'] +apartments['unit_1bed_2028_avg']+
apartments['unit_1bed_2029_avg'] + apartments['unit_1bed_2030_avg']) * 52

apartments['5yr_income_2bed'] = (apartments['unit_2bed_2026_avg'] + apartments['unit_2bed_2027_avg'] +apartments['unit_2bed_2028_avg']+
apartments['unit_2bed_2029_avg'] + apartments['unit_2bed_2030_avg']) * 52

apartments['5yr_income_3bed'] = (apartments['unit_3bed_2026_avg'] + apartments['unit_3bed_2027_avg'] +apartments['unit_3bed_2028_avg']+
apartments['unit_3bed_2029_avg'] + apartments['unit_3bed_2030_avg']) * 52

apartments['unit_avg_rental_income_5yr'] = (apartments['5yr_income_1bed']+apartments['5yr_income_2bed']+apartments['5yr_income_3bed']) / 3

apartments['unit_5yr_RETURN(%)'] = apartments['unit_avg_rental_income_5yr']/apartments['apartment_price']*100

apartments.sort_values(by='unit_5yr_RETURN(%)', ascending=False).head()
```

```python


houses['5yr_income_2bed'] = (houses['house_2bed_2026_avg'] + houses['house_2bed_2027_avg'] +houses['house_2bed_2028_avg']+
houses['house_2bed_2029_avg'] + houses['house_2bed_2030_avg']) * 52

houses['5yr_income_3bed'] = (houses['house_3bed_2026_avg'] + houses['house_3bed_2027_avg'] +houses['house_3bed_2028_avg']+
houses['house_3bed_2029_avg'] + houses['house_3bed_2030_avg']) * 52

houses['5yr_income_4bed'] = (houses['house_4bed_2026_avg'] + houses['house_4bed_2027_avg'] +houses['house_4bed_2028_avg']+
houses['house_4bed_2029_avg'] + houses['house_4bed_2030_avg']) * 52

houses['house_avg_rental_income_5yr'] = (houses['5yr_income_4bed']+houses['5yr_income_3bed']+houses['5yr_income_2bed']) / 3

houses['house_5yr_RETURN(%)'] = houses['house_avg_rental_income_5yr']/houses['House_price']*100

houses.sort_values(by='house_5yr_RETURN(%)', ascending=False).head(20)
```

```python
af = most_affordable.drop(columns=['livability_score', 'postcode']).rename(columns={'suburb':'Suburb'})

liv = most_livable.drop(columns=['affordability_score', 'postcode']).rename(columns={'suburb':'Suburb'})
```

```python
af.head()
```

```python
m = af.merge(liv, on='Suburb', how='left')

merged = m.merge(all, on='Suburb', how='left')

merged.head()
```

```python
show = merged.dropna()

keep = ['Suburb', 'livability_rank', 'affordability_rank', 'House_ALL_avg_price_change(%)', 'apartment_ALL_avg_price_change(%)', 'house_avg_price_today',
        'unit_avg_price_today', 'house_2bed_Rent_today', 'house_3bed_Rent_today', 'house_4bed_Rent_today']

drop = [col for col in show.columns if col not in keep]

dashboard = show.drop(columns=drop).rename(columns={'House_ALL_avg_price_change(%)' : 'House_avg_annual_rent_change(%)',
                                                    'apartment_ALL_avg_price_change(%)': 'Unit_avg_annual_rent_change(%)'})

dashboard = dashboard.sort_values(by='livability_rank', ascending=True).head(10)
```

```python
dashboard_pretty = dashboard.rename(columns={
    'livability_rank': 'Livability Rank',
    'affordability_rank': 'Affordability Rank',
    'House_avg_annual_rent_change(%)': 'House Rent Δ (%)',
    'Unit_avg_annual_rent_change(%)': 'Unit Rent Δ (%)', 
    'house_avg_price_today': 'House Avg Rent Today', 'unit_avg_price_today' : 'Unit Avg Rent Today'
})

styled = (dashboard_pretty
          .style
          .format({'House Rent Δ (%)': '{:.2f}', 'Unit Rent Δ (%)': '{:.2f}'})
          .hide(axis='index'))
styled


```

```python
styled.to_excel("../data/outputs/dashboard.xlsx")
```
