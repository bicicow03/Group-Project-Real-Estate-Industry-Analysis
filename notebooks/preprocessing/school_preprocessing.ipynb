```python
import pandas as pd

# Load the raw school location data from CSV file
df = pd.read_csv('../data/landing/schools/school_locations_2024.csv', encoding='latin-1')

```

```python
# Inspect the dataframe structure and data types
df.info()
```

```python
# Select relevant columns for further analysis
schools_df = df[['School_Name', 'School_Type', 'X', 'Y', 'Address_Town', 'Address_Postcode','Region_Name', 'LGA_Name', 'LGA_ID']]
```

```python
# Display the selected columns to verify selection
schools_df
```

```python
# Check for missing values and data types after column selection
schools_df.info()
```

```python
# Remove rows with missing values to ensure data quality
schools_df.dropna(inplace=True)
```

```python
# Rename coordinate columns for clarity
schools_df.rename(columns={'X': 'Longitude', 'Y': 'Latitude'}, inplace=True)
```

```python
# Display the cleaned and renamed dataframe
schools_df
```

```python
# create the processed data directory if it doesn't exist
import os
os.makedirs('../data/processed/schools', exist_ok=True)

# Save the processed school location data to a new CSV file
schools_df.to_csv('../data/processed/schools/school_locations.csv', index=False)
```
